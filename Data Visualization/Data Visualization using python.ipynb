```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_datascience = pd.read_csv('C:/Users/rubay/Downloads/Topic_Survey_Assignment.csv')

df_datascience
```

```python
df_sorted = df_datascience.sort_values(['Very interested'],ascending=False, inplace=True)

df_datascience
```

```python
df_datascience= df_datascience.astype(float)
df_datascience= (100*df_datascience/2233).round(2)

df_datascience
```

```python
color_list=['#5cb85c','#5bc0de', '#d9534f']

ax= df_datascience.plot(kind="bar",figsize=(20,8), width=0.8, color=color_list, fontsize=14)


plt.title('Percentage of respondents interested in data science areas',fontsize=16)


for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')

plt.show()
```
